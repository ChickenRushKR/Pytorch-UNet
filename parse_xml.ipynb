```python
with open("annotation2.xml", "r") as f:
    while True:
        l = f.readline()
        if not l: break
        find_image = l.find('<image')
        if find_image == -1:
            continue
        else:
            image = l[find_image:]
            parse_space = image.split(" ")
            name = parse_space[2][6:len(parse_space[2])-1]

            l = f.readline()
            parse_space = l.split(" ")
            points = (parse_space[8][8:len(parse_space[8])-1]).replace(";",",").split(",")
            print(name, points[0], points[1], points[2], points[3])
            
```
